```python
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
# fix random seed for reproducibility
numpy.random.seed(7)
```

```python
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
# split into input (X) and output (Y) variables
X = dataset[:,0:8] #pierwsze 7 kolumn
Y = dataset[:,8] # 8 kolumna
```

```python
train = pd.read_csv("pima-indians-diabetes.csv", delimiter=",", header=None)
```

```python
train
```

```python
X_train = (train.iloc[:,1:].values).astype('float32') # all pixel values
#Y_train = train.iloc[:,0].values.astype('int32') # only labels
#X_test = test.values.astype('int32')
```

```python
X_train=train.iloc[:,0:8]
```

```python
X_train
```

```python
X_test=train.iloc[:,8]
```

```python
X_test
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Fit the model
model.fit(X_train, X_test, epochs=50, batch_size=20)
```

```python
scores = model.evaluate(X_train, X_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# calculate predictions
predictions = model.predict(X_train)
predictions
```

```python
# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

```python
results = np.argmax(rounded,axis = 0)
results = pd.Series(rounded)
```

```python
submission = pd.concat([pd.Series(range(1,769)),results],axis = 1)
```
