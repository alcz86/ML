## Seasons

```python
games_athletes = df.pivot_table(df, index=['Games'], aggfunc=lambda x: len(x.unique())).reset_index()[['Games','ID']]
fig, ax = plt.subplots(figsize=(22,6))
a = sns.barplot(x='Games', y='ID', data=games_athletes, ax=ax, color="#2196F3")
a.set_xticklabels(labels=games_athletes['Games'],rotation=90)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')

ax.set_xlabel('Olympic Game', size=14, color="#0D47A1")
ax.set_ylabel('Number of Athletes', size=14, color="#0D47A1")
ax.set_title('Athletes in each Olympic game', size=18, color="#0D47A1")

plt.show()
```

```python
summer_athletes = df[df['Season']=='Summer'].pivot_table(df, index=['Games'], aggfunc=lambda x: len(x.unique())).reset_index()[['Games','ID']]

fig, ax = plt.subplots(figsize=(22,6))
a = sns.barplot(x='Games', y='ID', data=summer_athletes, ax=ax, color="palegreen")
a.set_xticklabels(labels=summer_athletes['Games'],rotation=90)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')

ax.set_xlabel('Summer Olympic Game', size=14, color="green")
ax.set_ylabel('Number of Athletes', size=14, color="green")
ax.set_title('Athletes in each Summer Olympic game', size=18, color="green")

plt.show()
```

```python
winter_athletes = df[df['Season']=='Winter'].pivot_table(df, index=['Games'], aggfunc=lambda x: len(x.unique())).reset_index()[['Games','ID']]

fig, ax = plt.subplots(figsize=(22,6))
a = sns.barplot(x='Games', y='ID', data=winter_athletes, ax=ax, color="lightblue")
a.set_xticklabels(labels=winter_athletes['Games'],rotation=90)

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')

ax.set_xlabel('Winter Olympic Game', size=14, color="darkblue")
ax.set_ylabel('Number of Athletes', size=14, color="darkblue")
ax.set_title('Athletes in each Winter Olympic game', size=18, color="darkblue")

plt.show()
```

## Poland

```python
data_team=data[data.Team=="Poland"]
data_pl=data_team.loc[:,["ID","Year"]]
```

```python
plMedals = data_team[pd.notna(data_team.Medal)]
```

```python
sorter=['Gold','Silver','Bronze']
```

```python
sorterIndex = dict(zip(sorter,range(len(sorter))))
plMedals['Tm'] = plMedals['Medal'].map(sorterIndex)
```

```python
plMedals.Tm = plMedals.Tm.astype("category")
plMedals.Tm.cat.set_categories(sorter, inplace=True)
#plMedals.drop('Tm', 1, inplace = True)
```
